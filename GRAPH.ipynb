```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x = np.arange(0,10)
y = np.arange(20,30)
```

```python
plt.scatter(x,y, c ='r', marker ='*')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('scatter plot Using Matplotlib')
plt.show()
```

```python
plt.plot(x,y, c='y')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('Line Graph')
plt.show()
```
