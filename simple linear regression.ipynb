# Cars Speed Data -In the example below, the x-axis represents age,
and the y-axis represents speed. We have registered the age and speed of 13 cars as they were passing a tollbooth. Let us see if the data we collected could be used in a linear regression:

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = [5,7,8,7,2,17,2,9,4,11,10,9,6]
y = [99,86,87,88,111,86,103,70,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
from scipy import stats
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
def myfunc(x):
  return slope * x + intercept
```

```python
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
